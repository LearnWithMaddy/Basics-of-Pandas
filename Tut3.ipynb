## How to sort Dataframes

```python
import pandas as pd
```

Sorting For Pandas DataFrame and series

```python
movies = pd.read_csv("http://bit.ly/imdbratings")
```

```python
movies.head()
```

```python
movies.title.sort_values()
```

```python
movies['title'].sort_values()
```

```python
type(movies['title'].sort_values())
```

```python
movies['title'].sort_values(ascending = False)
```

```python
movies.sort_values('title' , ascending = False)
```

```python
movies.sort_values(['duration' , 'star_rating'])
```
